* seperate users and messages

```python
users=[]
msg=[]

for message in df['user_name']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        # user name
        users.append(entry[1])
        msg.append(entry[2])
    else:
        users.append('group notification')
        msg.append(entry[0])

df['user']=users
df['message']=msg
df.drop(columns=['user_name'],inplace=True)
```

```python
#!/usr/bin/env python
import re
def emoji_count(msg):
        # text = u'Congratul💐ations  to all who got selected💐💐\n'
        # print(text) # with emoji

        emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                                "]+", flags=re.UNICODE)
        emoji=emoji_pattern.findall(msg) # no emoji
        # Split combined emojis into individual elements
        emoji = [char for emoji in emoji for char in emoji]
        return emoji
```

```python
chat_emojis=[]
for msg in df['message']:
    for i in msg.split():
        w=emoji_count(i)
        if len(w)!=0:
            chat_emojis.extend(w)
emoji_df=pd.DataFrame(Counter(chat_emojis).most_common(20)).reindex().rename(
    columns={
        0:"Emoji",
        1:"Count"
    }
)
emoji_df
```

* making period for of hours

```python
period=[]
for hour in df[['day_name','Hour']]['Hour']:
    if hour==23:
        period.append(str(hour)+"-"+str("00"))
    elif hour ==0:
        period.append(str('00')+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period']=period
```

```python
df
```

```python
df.pivot_table(index='day_name',columns='period',
                   values='message',aggfunc='count').fillna(0)
```

```python
import seaborn as sns

plt.figure(figsize=(18,5))
sns.heatmap(
    df.pivot_table(index='day_name',columns='period',
                   values='message',aggfunc='count').fillna(0),
    cmap='Greens'
)
plt.xticks(rotation='vertical')
plt.show()
```
